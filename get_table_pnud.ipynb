```python
import pandas as pd
import lxml
#libs necessárias para puxar a tabela
```

```python
files = pd.read_html("https://www.undp.org/pt/brazil/idhm-munic%C3%ADpios-2010")
```

```python
df = files[0] #A lib puxa todas as tabelas e transforma em lista. Pegamos a primeira - e única.
```

```python
df.head(10) #primeiras 10 linhas
```

```python
df[["Município", "Estado"]] = df['Município'].str.split("(", expand=True) #ajuste para separar as colunas de Estado
```

```python
df
```

```python
df["Estado"] = df['Estado'].str.replace(")", "") #retirando o parênteses que ficou.
```

```python
df.to_csv("pnud_data.csv", index=False, sep=";") #salvando em CSV
```
